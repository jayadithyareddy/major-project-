```python

# CREDIT CARD FRAUD DETECTION
# PART 1 - DATA LOADING, EDA & PREPROCESSING
# ===============================
# Import Libraries
# ===============================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
sns.set_style("whitegrid")
plt.rcParams["figure.figsize"]=(8,5)

# ===============================
# Load Dataset
# ===============================

df = pd.read_csv("/content/creditcard.csv")
print("="*60)
print("DATASET LOADED SUCCESSFULLY")
print("="*60)

# ===============================
# Basic Information
# ===============================

print(df.head())
print("\nDataset Shape")
print(df.shape)
print("\nColumns")
print(df.columns)
print("\nInformation")
print(df.info())
print("\nStatistical Summary")
print(df.describe())

# ===============================
# Missing Values
# ===============================
print("\nMissing Values")
print(df.isnull().sum())

# ===============================
# Duplicate Values
# ===============================
duplicates=df.duplicated().sum()
print("\nDuplicate Rows :",duplicates)

# ===============================
# Remove Duplicates
# ===============================

df=df.drop_duplicates()
print("Dataset Shape After Removing Duplicates")
print(df.shape)

# ==================================================
# CLASS DISTRIBUTION
# ==================================================

print(df['Class'].value_counts())
plt.figure(figsize=(6,5))
sns.countplot(x='Class',data=df,palette='Set2')
plt.title("Class Distribution")
plt.xlabel("0 = Genuine")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()

# ==================================================
# PIE CHART
# ==================================================

plt.figure(figsize=(6,6))
df.Class.value_counts().plot(
    kind='pie',
    autopct='%1.2f%%',
    colors=['skyblue','red'],
    labels=['Genuine','Fraud']
)

plt.title("Fraud vs Genuine Transactions")
plt.ylabel("")
plt.show()

# ==================================================
# TRANSACTION AMOUNT DISTRIBUTION
# ==================================================

plt.figure(figsize=(10,5))
sns.histplot(df['Amount'],bins=60,kde=True,color='green')
plt.title("Transaction Amount Distribution")
plt.show()

# ==================================================
# TIME DISTRIBUTION
# ==================================================

plt.figure(figsize=(10,5))
sns.histplot(df['Time'],bins=60,color='orange')
plt.title("Transaction Time Distribution")
plt.show()

# ==================================================
# BOXPLOT
# ==================================================

plt.figure(figsize=(8,5))
sns.boxplot(x='Class',y='Amount',data=df)
plt.title("Amount vs Class")
plt.show()

# ==================================================
# VIOLIN PLOT
# ==================================================

plt.figure(figsize=(8,5))
sns.violinplot(x='Class',y='Amount',data=df)
plt.title("Fraud vs Genuine Amount")
plt.show()

# ==================================================
# DISTRIBUTION OF AMOUNT
# ==================================================

plt.figure(figsize=(8,5))
sns.kdeplot(df['Amount'],fill=True)
plt.title("Amount Density")
plt.show()

# ==================================================
# DISTRIBUTION OF TIME
# ==================================================

plt.figure(figsize=(8,5))
sns.kdeplot(df['Time'],fill=True,color='orange')
plt.title("Time Density")
plt.show()

# ==================================================
# CORRELATION HEATMAP
# ==================================================

plt.figure(figsize=(18,14))
corr=df.corr()
sns.heatmap(
    corr,
    cmap='coolwarm',
    center=0
)

plt.title("Correlation Heatmap")
plt.show()

# ==================================================
# TOP POSITIVE CORRELATION
# ==================================================

corr_class=df.corr()['Class'].sort_values(ascending=False)
print("\nTop Positive Correlations")
print(corr_class.head(10))
print("\nTop Negative Correlations")
print(corr_class.tail(10))

# ==================================================
# SELECTED FEATURE DISTRIBUTION
# ==================================================

features=['V10','V12','V14','V17','Amount']

for feature in features:

    plt.figure(figsize=(8,4))
    sns.histplot(df[feature],kde=True)
    plt.title(feature)
    plt.show()

# ==================================================
# PAIRPLOT
# ==================================================

sample=df.sample(2000,random_state=42)
sns.pairplot(
    sample[['V10','V12','V14','Amount','Class']],
    hue='Class',
    corner=True
)
plt.show()

# ==================================================
# FEATURE SCALING
# ==================================================

scaler=StandardScaler()
df['Amount']=scaler.fit_transform(df[['Amount']])
df['Time']=scaler.fit_transform(df[['Time']])
print("\nScaling Completed")

# ==================================================
# TRAIN TEST SPLIT
# ==================================================

X=df.drop("Class",axis=1)
y=df["Class"]
X_train,X_test,y_train,y_test=train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y

)

print("\nTrain Shape :",X_train.shape)
print("Test Shape :",X_test.shape)
print("\nPart 1 Completed Successfully")
# ==========================================================
# PART 2 : MACHINE LEARNING MODELS
# ==========================================================

# Install required libraries (Colab)
!pip -q install imbalanced-learn xgboost
import xgboost as xgb
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix,
    roc_auc_score,
    roc_curve,
    precision_recall_curve,
    auc
)
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler

# ==========================================================
# LOGISTIC REGRESSION
# ==========================================================

print("="*60)
print("LOGISTIC REGRESSION")
print("="*60)
lr = LogisticRegression(max_iter=1000)

lr.fit(X_train, y_train)

y_pred_lr = lr.predict(X_test)

print(classification_report(y_test, y_pred_lr))

lr_acc = accuracy_score(y_test, y_pred_lr)

print("Accuracy :", lr_acc)

# ==========================================================
# RANDOM UNDER SAMPLING
# ==========================================================

print("="*60)
print("UNDER SAMPLING")
print("="*60)

rus = RandomUnderSampler(random_state=42)

X_under, y_under = rus.fit_resample(X_train, y_train)

lr_under = LogisticRegression(max_iter=1000)

lr_under.fit(X_under, y_under)

pred_under = lr_under.predict(X_test)

print(classification_report(y_test, pred_under))

# ==========================================================
# SMOTE
# ==========================================================

print("="*60)
print("SMOTE")
print("="*60)

smote = SMOTE(random_state=42)

X_smote, y_smote = smote.fit_resample(X_train, y_train)

lr_smote = LogisticRegression(max_iter=1000)

lr_smote.fit(X_smote, y_smote)

pred_smote = lr_smote.predict(X_test)

print(classification_report(y_test, pred_smote))

# ==========================================================
# RANDOM FOREST
# ==========================================================

print("="*60)
print("RANDOM FOREST")
print("="*60)

rf = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    n_jobs=-1
)

rf.fit(X_train, y_train)

pred_rf = rf.predict(X_test)

print(classification_report(y_test, pred_rf))

rf_acc = accuracy_score(y_test, pred_rf)

print("Accuracy :", rf_acc)

# ==========================================================
# XGBOOST
# ==========================================================

print("="*60)
print("XGBOOST")
print("="*60)

xgb_model = xgb.XGBClassifier(
    objective='binary:logistic',
    learning_rate=0.05,
    max_depth=6,
    n_estimators=300,
    subsample=0.8,
    colsample_bytree=0.8,
    eval_metric='logloss',
    random_state=42
)

xgb_model.fit(X_train, y_train)

pred_xgb = xgb_model.predict(X_test)

print(classification_report(y_test, pred_xgb))

xgb_acc = accuracy_score(y_test, pred_xgb)

print("Accuracy :", xgb_acc)

# ==========================================================
# GRID SEARCH CV
# ==========================================================

print("="*60)
print("GRID SEARCH CV")
print("="*60)

params = {
    'learning_rate':[0.05,0.1],
    'max_depth':[4,6],
    'n_estimators':[100,200]
}

grid = GridSearchCV(

    estimator=xgb.XGBClassifier(
        objective='binary:logistic',
        eval_metric='logloss',
        random_state=42
    ),

    param_grid=params,

    cv=3,

    scoring='roc_auc',

    n_jobs=-1

)

grid.fit(X_train, y_train)

print("Best Parameters")

print(grid.best_params_)

best_model = grid.best_estimator_

pred_best = best_model.predict(X_test)

print(classification_report(y_test, pred_best))

best_acc = accuracy_score(y_test, pred_best)

print("Best Accuracy :", best_acc)

# ==========================================================
# CONFUSION MATRIX
# ==========================================================

cm = confusion_matrix(y_test, pred_best)

plt.figure(figsize=(6,5))

sns.heatmap(
    cm,
    annot=True,
    cmap="Blues",
    fmt="d"
)

plt.title("Confusion Matrix")

plt.xlabel("Predicted")

plt.ylabel("Actual")

plt.show()

# ==========================================================
# ROC CURVE
# ==========================================================

prob = best_model.predict_proba(X_test)[:,1]
fpr, tpr, _ = roc_curve(y_test, prob)
roc = roc_auc_score(y_test, prob)
plt.figure(figsize=(7,5))
plt.plot(fpr, tpr, label="AUC = %.4f"%roc)
plt.plot([0,1],[0,1],'r--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.grid(True)
plt.show()
print("ROC AUC :", roc)

# ==========================================================
# PRECISION RECALL CURVE
# ==========================================================

precision, recall, _ = precision_recall_curve(y_test, prob)
pr_auc = auc(recall, precision)
plt.figure(figsize=(7,5))
plt.plot(recall, precision, color="purple")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision Recall Curve")
plt.grid(True)
plt.show()
print("PR AUC :", pr_auc)

# ==========================================================
# FEATURE IMPORTANCE
# ==========================================================

plt.figure(figsize=(10,6))
xgb.plot_importance(
    best_model,
    max_num_features=15
)

plt.title("Top 15 Important Features")
plt.show()

# ==========================================================
# MODEL COMPARISON
# ==========================================================

result = pd.DataFrame({

    "Model":[
        "Logistic Regression",
        "Random Forest",
        "XGBoost"
    ],

    "Accuracy":[
        lr_acc,
        rf_acc,
        best_acc
    ]

})

print(result)

plt.figure(figsize=(8,5))

sns.barplot(
    x="Model",
    y="Accuracy",
    data=result,
    palette="viridis"
)

plt.ylim(0.95,1)
plt.title("Model Accuracy Comparison")
plt.show()

print("="*60)
print("PART 2 COMPLETED SUCCESSFULLY")
print("="*60)
# ==========================================================
# PART 3 : ADVANCED VISUALIZATION & BUSINESS REPORT
# ==========================================================

from sklearn.metrics import precision_score, recall_score, f1_score
from sklearn.model_selection import learning_curve

# ==========================================================
# 1. ROC CURVE COMPARISON
# ==========================================================

models = {
    "Logistic Regression": lr,
    "Random Forest": rf,
    "XGBoost": best_model
}

plt.figure(figsize=(8,6))

for name, model in models.items():

    probability = model.predict_proba(X_test)[:,1]
    fpr, tpr, _ = roc_curve(y_test, probability)
    score = roc_auc_score(y_test, probability)
    plt.plot(fpr, tpr, linewidth=2,
    label=f"{name} (AUC={score:.4f})")

plt.plot([0,1],[0,1],'k--')

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve Comparison")
plt.legend()
plt.grid(True)
plt.show()


# ==========================================================
# 2. PRECISION RECALL F1 SCORE COMPARISON
# ==========================================================

comparison = pd.DataFrame({

    "Model":[
        "Logistic Regression",
        "Random Forest",
        "XGBoost"
    ],

    "Precision":[
        precision_score(y_test,y_pred_lr),
        precision_score(y_test,pred_rf),
        precision_score(y_test,pred_best)
    ],

    "Recall":[
        recall_score(y_test,y_pred_lr),
        recall_score(y_test,pred_rf),
        recall_score(y_test,pred_best)
    ],

    "F1 Score":[
        f1_score(y_test,y_pred_lr),
        f1_score(y_test,pred_rf),
        f1_score(y_test,pred_best)
    ]

})

print(comparison)

comparison.set_index("Model").plot(
    kind="bar",
    figsize=(10,6)
)

plt.title("Precision Recall F1 Comparison")
plt.xticks(rotation=0)
plt.grid(True)
plt.show()


# ==========================================================
# 3. FEATURE IMPORTANCE
# ==========================================================

importance = pd.DataFrame({

    "Feature":X.columns,

    "Importance":best_model.feature_importances_

})

importance = importance.sort_values(
    by="Importance",
    ascending=False
)

plt.figure(figsize=(10,7))

sns.barplot(
    data=importance.head(15),
    x="Importance",
    y="Feature",
    palette="rocket"
)

plt.title("Top 15 Important Features")

plt.show()


# ==========================================================
# 4. LEARNING CURVE
# ==========================================================

train_sizes, train_scores, test_scores = learning_curve(

    estimator=best_model,

    X=X,

    y=y,

    cv=5,

    scoring="accuracy",

    train_sizes=np.linspace(0.1,1.0,8),

    n_jobs=-1

)

train_mean = train_scores.mean(axis=1)

test_mean = test_scores.mean(axis=1)

plt.figure(figsize=(8,5))

plt.plot(train_sizes,train_mean,
         marker='o',
         label="Training Accuracy")

plt.plot(train_sizes,test_mean,
         marker='o',
         label="Validation Accuracy")

plt.xlabel("Training Samples")

plt.ylabel("Accuracy")

plt.title("Learning Curve")

plt.legend()

plt.grid(True)

plt.show()


# ==========================================================
# 5. MODEL ACCURACY GRAPH
# ==========================================================

accuracy_df = pd.DataFrame({

    "Model":[
        "Logistic Regression",
        "Random Forest",
        "XGBoost"
    ],

    "Accuracy":[
        lr_acc,
        rf_acc,
        best_acc
    ]

})

plt.figure(figsize=(8,5))

sns.barplot(
    data=accuracy_df,
    x="Model",
    y="Accuracy",
    palette="viridis"
)

plt.ylim(0.95,1)

for index,value in enumerate(accuracy_df["Accuracy"]):

    plt.text(index,value+0.0003,
             f"{value:.4f}",
             ha='center',
             fontsize=11)

plt.title("Accuracy Comparison")

plt.show()


# ==========================================================
# 6. FINAL RESULTS TABLE
# ==========================================================

results = pd.DataFrame({

"Model":[
"Logistic Regression",
"Random Forest",
"XGBoost"
],

"Accuracy":[
lr_acc,
rf_acc,
best_acc
],

"ROC AUC":[
roc_auc_score(y_test,lr.predict_proba(X_test)[:,1]),
roc_auc_score(y_test,rf.predict_proba(X_test)[:,1]),
roc_auc_score(y_test,best_model.predict_proba(X_test)[:,1])
]

})

print("\n")
print("="*70)
print("FINAL MODEL COMPARISON")
print("="*70)

print(results)


# ==========================================================
# 7. BUSINESS IMPACT
# ==========================================================

print("\n")
print("="*70)
print("BUSINESS IMPACT")
print("="*70)

print("""
• Detects fraudulent transactions automatically.

• Minimizes financial losses.

• Improves customer trust.

• Enables banks to monitor suspicious activities.

• Supports real-time fraud detection.

• Reduces manual verification effort.

• XGBoost achieved the best performance
  and is selected as the final deployment model.
""")


# ==========================================================
# 8. CONCLUSION
# ==========================================================

print("\n")
print("="*70)
print("CONCLUSION")
print("="*70)

print("""
The project successfully identified fraudulent
transactions using Machine Learning algorithms.

Among all models,
XGBoost achieved the highest Accuracy and ROC-AUC,
making it the best model for fraud detection.
""")


# ==========================================================
# 9. FUTURE SCOPE
# ==========================================================

print("\n")
print("="*70)
print("FUTURE SCOPE")
print("="*70)

print("""
1. Deploy using Flask or Streamlit.

2. Integrate real-time payment gateway.

3. Apply Deep Learning models.

4. Use AutoML for optimization.

5. Deploy on Cloud (AWS/Azure/GCP).

6. Build a real-time fraud detection dashboard.
""")


print("\n")
print("="*70)
print("🎉 PROJECT COMPLETED SUCCESSFULLY 🎉")
print("="*70)
```
